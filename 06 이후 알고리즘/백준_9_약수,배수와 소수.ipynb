```python
M = int(input())
N = int(input())

def is_prime(n):
    tot=0
    if n ==1:
        return 0
    else:
        for i in range(2,int(n**(1/2)+1)):
            if n%i==0:
                tot+=1
        if tot ==0:
            return n
        else:
            return 0
ans = 0
ans_min = N
for i in range(M,N+1):
    ans+=is_prime(i)
    if is_prime(i)==i:
        ans_min = min(is_prime(i),ans_min)
if ans >0:
    print(ans)
    print(ans_min)
else:
    print(-1)
```
