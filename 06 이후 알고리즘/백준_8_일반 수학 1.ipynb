11005 진법 변환 2


```python
numList = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"
a,b = map(int,input().split())

answer = ''

while a:
    answer+=str(numList[a%b])
    a//=b    
print(answer[::-1])
```

2720 세탁소 사장 동혁


```python
Q,D,N,P = 25,10,5,1
a_list = []
T = int(input())
for i in range(T):
    a_list = []
    a = int(input())
    a//Q
    a_list.append(a//Q)
    a-=a//Q *Q
    a_list.append(a//D)
    a-=a//D *D
    a_list.append(a//N)
    a-=a//N *N
    a_list.append(a//P)
    print(*a_list)
```

2292 벌집


```python
# 1 | 2-7 | 8-19|20-37|38-61|
# 1 | 6 | 12 | 18 | 24 |

tot = 1
d=1
n = int(input())
while True:
    if n <= tot:
        break
    tot+=6*d
    d+=1
print(d)
```

1193 분수 찾기


```python
# 1,1 -> 1,2 -> 2,1 -> 3,1 -> 2,2 -> 1,3 -> 1,4 -> 2,3 -> 3,2 -> 4,1 -> 5,1
# 1 2 3 4 5
d = 1
ans = 1
n= int(input())
while True:
    if n <= d:
        break
    n-=d
    d+=1

if d%2==0:
    print(n,'/',d+1-n,sep='')
else:
    print(d+1-n,'/',n,sep='')
```

2869 달팽이는 올라가고 싶다


```python
A,B,V = map(int,input().split())

n = (V-A)/(A-B)+1
a = int(n)
if n ==a:
    print(a)
else:
    print(a+1)
```
