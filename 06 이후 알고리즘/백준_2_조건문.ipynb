### 1330 두 수 비교하기


```python
A,B = map(int,input().split())

if A>B:
    print('>')
elif A<B:
    print('<')
else:
    print('==')
```

14681 사분면 고르기


```python
x = int(input())
y = int(input())

if x<0:
    if y<0:
        print(3)
    else:
        print(2)
else:
    if y<0:
        print(4)
    else:
        print(1)
```

2884 알람 시계


```python
H,M = map(int,input().split())

H_list =[]
M_list =[]

for i in range(24):
    H_list.append(i)
for j in range(60):
    M_list.append(j)
    
if M>=45:
    print(H,(M-45))
else:
    print(H_list[H_list.index(H)-1], M_list[-(M_list.index(45-M))])
```

2480 주사위 세개


```python
A,B,C = map(int,input().split())

if A==B==C:
    print(10000+A*1000)
elif A==B or B==C:
    print(1000+B*100)
elif A==C:
    print(1000+A*100)
else:
    print(max(A,B,C)*100)
```
