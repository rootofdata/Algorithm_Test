```python
N,M = map(int,input().split())
a_list = list()
for i in range(N):
    a_list.append(i+1)
for _ in range(M):
    i,j = map(int,input().split())
    a_list[i-1],a_list[j-1] = a_list[j-1],a_list[i-1]
print(*a_list)
```
