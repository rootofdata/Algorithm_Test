10809 알파벳 찾기


```python
S = input()
a = 'abcdefghijklmnopqrstuvwxyz'
a_list = [-1]*24

for i in range(len(S)):
    if S[i] in a:
        if a_list[a.index(S[i])] == -1:
            a_list[a.index(S[i])] = i
print(*a_list)
```

5622 다이얼


```python
N = input()
a_list = ['ABC','DEF','GHI','JKL','MNO','PQRS','TUV','WXYZ']
tot=0
for i in range(len(N)):
    for j in a_list:
        if N[i] in j:
            tot+= a_list.index(j)+3
print(tot)
```
