2738 행렬 덧셈


```python
N,M = map(int,input().split())
A_list = list()
B_list = list()

for i in range(N):
    A_list.append(list(map(int,input().split())))
for i in range(N):
    B_list.append(list(map(int,input().split())))
tot_list = []
for i in range(N):
    for j in range(M):
        tot_list.append(A_list[i][j]+B_list[i][j])
for i in range(N):
    print(*tot_list[i*M:i*M+M])
```

2566 최댓값


```python
a_list =[]
for i in range(9):
    a_list.append(list(map(int,input().split())))
tot = 0
row = 0
for i in range(9):
    if tot <= max(a_list[i]):
        row = i+1
    tot = max(tot,max(a_list[i]))
    
column = a_list[row-1].index(tot)+1
print(tot)
print(row,column)
```

10798 세로읽기


```python
a_list = []

for i in range(5):
    a_list.append(list(input()))
for j in range(15):
    for i in range(5):
        if j < len(a_list[i]):
            print(a_list[i][j],end='')
```

2563 색종이


```python
m = int(input())
l_list = [[0 for i in range(101)] for i in range(101)]
for _ in range(m):
    a,b = map(int,input().split())
    for i in range(10):
        for j in range(10):
            l_list[a+i][b+j] = 1
count = 0
for i in range(101):
    count+=sum(l_list[i])
print(count)
```
