```python
N = int(input())

a_list = []
for i in range(N):
    a_list.append(list(map(int,input().split())))
b_list = []
cnt = 0
plus_list = 0
minus_list = 0
zero_list = 0

def check(x,y,N):
    global a_list,plus_list,minus_list,zero_list
    
    C = a_list[x][y]
    for i in range(x,x+N):
        for j in range(y,y+N):
            if a_list[i][j] != C:
                for k in range(3):
                    for m in range(3):
                        check(x+k*N//3,y+m*N//3,N//3)
                return
    if C==0:
        zero_list+=1
    elif C==1:
        plus_list +=1
    elif C==-1:
        minus_list +=1

check(0,0,N)

print(minus_list)
print(zero_list)
print(plus_list)
```
