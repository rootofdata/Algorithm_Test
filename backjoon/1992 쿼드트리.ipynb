```python
N = int(input())
a_list=[]
for i in range(N):
    a_list.append(list(input()))
ans = ''

def check(x,y,N):
    global a_list,ans
    C=0
    for i in range(x,x+N):
        for j in range(y,y+N):
            C+=int(a_list[i][j])
    if not C:
        ans+='0'
    elif C==N**2:
        ans+='1'
    else:
        N=N//2
        ans+='('
        check(x,y,N)
        check(x,y+N,N)
        check(x+N,y,N)
        check(x+N,y+N,N)
        ans+=')'
    return
check(0,0,N)
print(ans)
```
