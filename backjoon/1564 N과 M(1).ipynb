```python
N,M = map(int,input().split())

L1=[]
def Recursive(m):
    if 1<=m:
        for i in range(1,N+1):
            if i not in L1:
                L1.append(i)
                Recursive(m-1)
                L1.pop()
    else:
        print(*L1)
                
                            
Recursive(M)    
```

```python
N,M = map(int,input().split())

from itertools import permutations as per
numbers = list(range(1,N+1))
perm = list(per(numbers,M))
for i in perm:
    print(*i)
```
