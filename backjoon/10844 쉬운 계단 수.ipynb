```python
N = int(input())

a_list = [i for i in range(10**(N-1),10**N)]

def ans(N):
    N = str(N)
    tot=0
    ans =0
    for i in range(1,len(N)):
        if abs(int(N[i]) - int(N[i-1])) ==1:
            tot+=1
    if tot==len(N)-1:
        ans=1
    return ans
res=0
for i in range(len(a_list)):
    res+=ans(a_list[i])
print(res)
```

```python
N = int(input())
result =0
def ans(N):
    N = str(N)
    tot=0
    ans =0
    for i in range(1,len(N)):
        if abs(int(N[i]) - int(N[i-1])) ==1:
            tot+=1
    if tot==len(N)-1:
        ans=1
    return ans

for i in range(10**(N-1),10**N):
    if ans(i)==1:
        result +=1
        
print(result% 1000000000)
```

```python
n = int(input())
dp = [[0 for i in range(10)] for j in range(101)]
for i in range(1, 10):
    dp[1][i] = 1
for i in range(2, n + 1):
    for j in range(10):
        if j == 0:
            dp[i][j] = dp[i - 1][1]
        elif j == 9:
            dp[i][j] = dp[i - 1][8]
        else:
            dp[i][j] = dp[i - 1][j - 1] + dp[i - 1][j + 1]
print(sum(dp[n]) % 1000000000)
```

```python
N = int(input())

Num = [[0]*10 for _ in range(101)]
for i in range(10):
    Num[1][i] = 1

def Calc(Iter):
    for i in range(10):
        if 0 < i:
            Num[Iter][i] += Num[Iter-1][i-1]
        if i < 9:
            Num[Iter][i] += Num[Iter-1][i+1]

for i in range(2,101):
    Calc(i)

print(sum(Num[N][1:10])%1000000000)
```
