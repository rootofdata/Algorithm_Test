```python
N = int(input())
N_list = list(map(int,input().split()))
M = int(input())
M_list = list(map(int,input().split()))

card = dict()

for i in N_list:
    if i in card:
        card[i]+=1
    else:
        card[i] = 1
for i in M_list:
    print(card[i] if i in card else 0,end=' ')
```
