```python
from collections import deque

N = int(input())
ans = deque()
for _ in range(N):
    
    a = input().split()
  
    if a[0] == "push":
        ans.append(a[1])
    elif a[0] == "pop":
        print(ans.popleft() if ans else -1)
    elif a[0] == "size":
        print(len(ans))
    elif a[0] == "empty":
        print(1 if not ans else 0)
    elif a[0] == "front":
        print(ans[0] if ans else -1)
    else:
        print(ans[-1] if ans else -1)
```
