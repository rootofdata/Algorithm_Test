```python
from math import gcd
n = int(input())

def prime_number(n):
    if n <=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

for _ in range(n):
    ans = int(input())
    while not prime_number(ans):
        ans+=1
    print(ans)
    
```
