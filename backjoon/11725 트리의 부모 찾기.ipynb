```python
import sys
sys.setrecrusionlimit(10**6)
input = sys.stdin.readline

n = int(input())
graph = [[] for i in range(n+1)]
ans = [0 for i in range(n+1)]
visited = [False for i in range(n+1)]

for i in range(n-1):
    a,b = map(int,input().split())
    graph[a].append(b)
    graph[b].append(a)

def dfs(p):
    for i in graph[p]:
        if visited[i] == False:
            ans[i]=p
            visited[i]=True
            dfs(i)

visited[1] = True
dfs(1)

for i in range(2,n+1):
    print(ans[i])
```
