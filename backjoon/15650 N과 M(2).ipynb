```python
N,M = map(int,input().split())

from itertools import permutations as per

numbers = list(range(1,N+1))
perm = list(per(numbers,M))
b=set()
for i in range(len(perm)):
    perm[i]=sorted(perm[i])
    a=''
    for j in perm[i]:
        a=a+str(j)+' '
    b.add(a)

b=sorted(list(b))

for i in b:
    print(i)
#for i in perm:
#    print(*i)
```

```python
N,M = map(int,input().split())

L1=[]
def Recursive(m):
#    print('-',L1)
    if len(L1)<M:
        for i in range(m,N+1):
            if i not in L1:
                L1.append(i)
                Recursive(i+1)
                L1.pop()
    else:
        print(*L1)
                            
Recursive(1)    
```
