```python
# 시간초과

M = int(input())
N = int(input())
answer = 0
ans_min = N
def prime(n):
        ans =[]
        for i in range(1,n+1):
            if n%i==0:
                ans.append(i)
        if len(ans)==2:
            return n
        else : return 0
for i in range(M,N+1):
    answer+=prime(i)
    if prime(i)==i:
        ans_min = min(prime(i),ans_min)
print(answer)
print(ans_min)
```

```python
## 수정

M = int(input())
N = int(input())
answer = 0
ans_min = N
def prime(n):
        ans =[]
        if n==1:
            return 0
        else :
            for i in range(2,int(n**(1/2)+1)):
                if n%i==0:
                    ans.append(i)
            if len(ans)==0:
                return n
            else : return 0
for i in range(M,N+1):
    answer+=prime(i)
    if prime(i)==i:
        ans_min = min(prime(i),ans_min)

if answer>0:
    print(answer)
    print(ans_min)
else : print(-1)
```
