```python
T = int(input())

def recursion(s,l,r):
    if l>=r:
        return 1,l+1
    elif s[l]!=s[r]:
        return 0,l+1
    else:
        return recursion(s,l+1,r-1)
def ispalindrome(s):
    return recursion(s,0,len(s)-1)

for i in range(T):
    S = input()
#    if S[::-1]==S:
#        print(1,end='')
#    else :
#        print(0,end='')
    print(ispalindrome(S)[0],ispalindrome(S)[1])
    
```
