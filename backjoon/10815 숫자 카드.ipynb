```python
N = int(input())
a = list(map(int,input().split()))
M = int(input())
b = list(map(int,input().split()))
ans = [0]*M
for i in range(M):
    if b[i] in a:
        ans[i]=1
print(ans)
```

```python
N = int(input())
a = set(map(int,input().split()))
M = int(input())
b = list(map(int,input().split()))


s1 = set(b)
s = a&s1
result = []

for i in range(M):
    result.append(1 if b[i] in s else 0)
for i in range(len(result)):
    print(result[i],end=' ')
```
