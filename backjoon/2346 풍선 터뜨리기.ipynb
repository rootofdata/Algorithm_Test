```python
N = int(input())
from collections import deque
b = [i for i in range(1,N+1)]
a =deque(map(int,input().split()))
b = deque(i for i in range(1,N+1))

res= []

while len(a)>0:
    next = a.popleft()
    if next>0:
        next = -next+1
    else: next = -next
    a.rotate(next)
    res.append(b.popleft())
    b.rotate(next)
print(*res)
```
