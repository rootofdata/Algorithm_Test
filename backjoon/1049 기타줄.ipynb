```python
N,M = map(int,input().split())
ans = []
for i in range(M):
    ans.append(list(map(int,input().split())))
package=[]
price = []
for i,j in ans:
    package.append(i)
    price.append(j)
pack = min(min(package),min(price)*6)
answer=0
while True:
    if N>=6:
        answer +=pack
        N-=6
    else:
        if pack<=N*min(price):
            answer +=pack
            break
        else:
            answer+=N*min(price)
            break
print(answer)
```
