```python
T= int(input())

prime = [1 for i in range(T+1)]

prime[0]=prime[1]=0

for i in range(2,T):
    for j in range(i,T,i):
        if prime[j] == 0:
            prime[j] =1
        else:
            prime[j]=0
            
        
print(sum(prime))
```
