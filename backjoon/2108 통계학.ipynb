```python
N = int(input())
a=[]
for i in range(N):
    a.append(int(input()))
a.sort()
print(int(round(sum(a)/N,0)))
print(a[N//2])

from collections import Counter
c=Counter(a).most_common(2)
if len(a)>1:
    if c[0][1]==c[1][1]:
        print(c[1][0])
    else:
        print(c[0][0])
else:
    print(c[0][0])
print(max(a)-min(a))
```
