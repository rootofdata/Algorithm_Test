```python
T = int(input())
for _ in range(T):
    A,B = map(int,input().split())
#    a,b = max(A,B),min(A,B)
    answer = 0
    for i in range(min(A,B),A*B+1):
        if i%A==0 and i%B==0:
            answer = i
            break
    print(answer)
```
