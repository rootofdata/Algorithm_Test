```python
N = int(input())

a = list(map(int,input().split()))
b = list(map(int,input().split()))

max_ans=-1e10
min_ans= 1e10
    
def recur(add,sub,mul,div,n,res):
    global min_ans,max_ans
    if n<N-1:
        if 0< add:
            recur(add-1,sub,mul,div,n+1,res+a[n+1])
        if 0< sub:
            recur(add,sub-1,mul,div,n+1,res-a[n+1])
        if 0< mul:
            recur(add,sub,mul-1,div,n+1,res*a[n+1])
        if 0< div:
            recur(add,sub,mul,div-1,n+1,int(res/a[n+1]))
    else:
        max_ans = max(res,max_ans)
        min_ans = min(res,min_ans)

recur(b[0],b[1],b[2],b[3],0,a[0])
print(int(max_ans))
print(int(min_ans))
```
