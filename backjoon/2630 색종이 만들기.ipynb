```python
N = int(input())
a_list=[]
blue = 0
white = 0
for i in range(N):
    a_list.append(list(map(int,input().split())))
def check(x,y,N):
    global a_list,blue,white
    first = a_list[x][y]
    for i in range(x,x+N):
        for j in range(y,y+N):
            if first !=a_list[i][j]:
                check(x,y,N//2)
                check(x+N//2,y,N//2)
                check(x,y+N//2,N//2)
                check(x+N//2,y+N//2,N//2)
                return
    if first ==0:
        white +=1
    else:
        blue +=1
check(0,0,N)
print(white)
print(blue)
```
