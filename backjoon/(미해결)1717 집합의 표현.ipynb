```python
N,M = map(int,input().split())

a_list = [[0] for i in range(N+1)]

for i in range(M):
    a,b,c = map(int,input().split())
    if a==0:
        a_list[b].append(c)
        a_list[c].append(b)
    else:
        if b in a_list[c] or c in a_list[b]:
            print('YES')
        else:
            print('NO')
```
