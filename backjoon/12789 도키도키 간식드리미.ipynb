```python
from collections import deque

n = int(input())

row = deque(map(int,input().split()))
col = []

order = 1
while True:
    if row and row[0]==order:
        row.popleft()
        order+=1
    elif col and col[-1]==order:
        col.pop()
        order+=1
    elif row:
        pop = row.popleft()
        col.append(pop)
    else:
        break
if row or col:
    print('Sad')
else:
    print("Nice")    
```
