```python
N,M = map(int,input().split())
N_list = []
M_list = []
for i in range(N):
    N_list.append(list(map(int,input().split())))
for j in range(N):
    M_list.append(list(map(int,input().split())))
result = []
for i in range(N):
    for j in range(M):
        result.append(N_list[i][j]+M_list[i][j])
for i in range(N):
    print(*result[i*M:i*M+M])
```
