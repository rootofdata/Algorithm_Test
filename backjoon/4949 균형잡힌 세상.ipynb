```python
def aaa(a,n,m):
    ans=0
    for i in a:
        if i == n:
            ans+=1
        elif i==m:
            ans-=1
        if ans <0:
            break
    if ans==0:
        return 1
    else:
        return 0

while True:
    a = list(str(input()))
    ans=0
    ans += aaa(a,'(',')')
    ans += aaa(a,'[',']')
    ans += aaa(a,'[',']')
    ans += aaa(a,'[',']')
    if ans ==4:
        print('yes')
    else:
        print('no')
    if len(a)==1:
        break
```

```python
L1 = []
while(True):
    L1.append(input())
    if L1[-1] == '.':
        L1.pop()
        break

for k in L1:
    stack = []
    for i in k:
        if i == ('(') or i == ('['):
            stack.append(i)
        elif i == (')') or i == (']'):
            if len(stack) == 0:
                stack.append(i)
                break
            elif stack[-1] == '(' and i == ')':
                stack.pop()
            
            elif stack[-1] == '[' and i == ']':
                stack.pop()
            else:
                break
    if stack:
        print('no')
    else:
        print('yes')
```
