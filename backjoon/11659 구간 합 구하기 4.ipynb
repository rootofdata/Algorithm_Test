```python
N,M = map(int,input().split())
N_list = list(map(int,input().split()))

Sum = [0]*(N+1)
Temp = 0

for i in range(1,N+1):
    Temp+=N_list[i-1]
    Sum[i]=Temp
    
for _ in range(M):
    i,j = map(int,input().split())
    print(Sum[j]-Sum[i-1])
```
