```python
n = int(input())

def fib(n):
    if n==1 or n==2:
        return 1
    else:
        return fib(n-1)+fib(n-2)

def fibonachi(n):
    cnt=0
    f=[]
    f.append(1)
    f.append(1)
    for i in range(2,n+1):
        a = f[i-1]+f[i-2]
        f.append(a)
        cnt+=1
    return cnt-1
print(fib(n),fibonachi(n))
```
