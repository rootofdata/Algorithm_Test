```python
a,b,c,d,e,f = map(int,input().split())

if a*e !=d*b:
    x = int((c*e -f*b) // ((a*e)-(d*b)))
    if b!=0:
        y = int((c-a*x)//b)
    else:
        y = int((f-d*x)//e)

print(x,y)
```
