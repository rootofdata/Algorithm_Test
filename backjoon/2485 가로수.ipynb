```python
N = int(input())
b_list = []
b=0
for i in range(N):
    a = int(input())
    b = a-b
    b_list.append(b)
    b = a
import math
ans = 1000000
for i in range(len(b_list)):
    for j in range(i+1,len(b_list)):
        gcd = math.gcd(b_list[i],b_list[j])
        ans = min(ans,gcd)
tot = 0        
for i in b_list:
    tot = i//ans
print(tot)
```

```python
from math import gcd

N = int(input())
a = int(input())
arr = []

for i in range(N-1):
    num = int(input())
    arr.append(num - a)
    a = num

g = arr[0]
for j in range(1, len(arr)):
    g = gcd(g, arr[j])

result = 0

for each in arr:
    result += each // g - 1
print(result)
```
