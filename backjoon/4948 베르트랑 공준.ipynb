```python
#  

def prime_number(n):
    if n <=1:
        return 0
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return 0
    return n

while True:
    n = int(input())
    if n==0:
        break
    count=0
    for i in range(n+1,2*n+1):
        if prime_number(i)==i:
            count+=1
    print(count)
```

```python
# prime_list  ,     --> .

def prime_number(n):
    if n <=1:
        return 0
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return 0
    return n
prime_list=[]
for i in range(2,123456*2+1):
    if prime_number(i)==i:
        prime_list.append(i)
        
while True:
    n = int(input())
    if n==0:
        break
    count=0
    for i in prime_list:
        if i>2*n:
            break
        elif n<i<=2*n:
            count +=1
    print(count)
```
