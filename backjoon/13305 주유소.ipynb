```python
N = int(input())

dist = list(map(int,input().split()))
price = list(map(int,input().split()))
Tot = price[0]*dist[0]
curprice =0
for i in range(N-2):
    curprice = min(price[i],price[i+1])
    Tot = Tot + dist[i+1]*curprice
    
print(Tot)
```

```python
N = int(input())

dist = list(map(int,input().split()))
price = list(map(int,input().split()))
Tot = price[0]*dist[0]
m = price[0]

road = 0
for i in range(1,N-1):
    if price[i] <m:
        Tot += m*road
        road = dist[i]
        m = price[i]
    else:
        road +=dist[i]
    if i==N-2:
        Tot += m*road
        
print(Tot)
```
