```python
N = int(input())
s=[]
for i in range(N):
    li = list(map(int,input().split()))
    if li[0]==1:
        s.append(li[1])
    elif li[0]==2:
        print(s.pop() if len(s) else -1)
    elif li[0]==3:
        print(len(s))
    elif li[0]==4:
        print(0 if len(s) else 1)
    else:
        print(s[-1] if len(s) else -1)
```
