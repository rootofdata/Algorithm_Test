```python
row , col= map(int, input().split())

org = []
ans = []
for i in range (row):
    org.append([])
    temp = str(input())
    for j in range(col):
        org[i].append(temp[j])
        
wht = [['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],]
blk = [['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],['B','W','B','W','B','W','B','W'],['W','B','W','B','W','B','W','B'],]

def cmp(i,j,org,cmp):
    temp = 0
    for a in range(8):
        for b in range(8):
            if org[a+i][b+j]!=cmp[a][b]:
                temp +=1
    return temp

   
for i in range(0,row-7):
    for j in range(0,col-7):
        ans.append(min(cmp(i,j,org,wht),cmp(i,j,org,blk)))
        
print(min(ans))
```
