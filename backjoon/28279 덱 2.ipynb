```python
from collections import deque

N = int(input())
b = deque()
for _ in range(N):
    a = list(map(int,input().split()))
    if a[0]==1:
        b.appendleft(a[1])
    elif a[0]==2:
        b.append(a[1])
    elif a[0]==3:
        print(b.popleft() if b else -1)
    elif a[0]==4:
        print(b.pop() if b else -1)
    elif a[0]==5:
        print(len(b))
    elif a[0]==6:
        print(0 if b else 1)
    elif a[0]==7:
        print(b[0] if b else -1)
    else:
        print(b[-1] if b else -1)
        
```
