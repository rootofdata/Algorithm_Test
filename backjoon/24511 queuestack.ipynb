```python
from collections import deque
N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))
que = deque()
for i in range(N):
    if A[i]==0:
        que.append(B[i])

M = int(input())
C = list(map(int,input().split()))
for i in C:
    que.appendleft(i)
    print(que.pop(),end=' ')
```
