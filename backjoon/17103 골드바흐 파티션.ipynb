```python
def prime(n):
    if n<=1:
        return 0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return 0
    return n    
T = int(input())
prime_list = []
for i in range(2,1000000):
    if prime(i)==i:
        prime_list.append(i)

    
for i in range(T):
    N = int(input())
    ans =0
    for j in range(N//2+1):
        if j in prime_list and N-j in prime_list:
            ans+=1
    print(ans)
```

```python
def prime(n):
    if n<=1:
        return 0
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return 0
    return n    
T = int(input())
prime_list = []
    
for _ in range(T):
    N = int(input())
    ans =0
    for i in range(2,N):
        if prime(i)==i:
            prime_list.append(i)
        
    for j in range(N//2+1):
        if j in prime_list and N-j in prime_list:
            ans+=1
    print(ans)
```

```python
prime = [1 for i in range(1000001)]
prime[0] = prime[1] = 0

for i in range(2, 1000001):
    for j in range(2*i, 1000001, i):
        prime[j] = 0

T = int(input())

for i in range(T):    
    num = int(input())
    count = 0

    for i in range(2, num//2 + 1):
        if prime[i] and prime[num-i]:
            count += 1
    print(count)
```
