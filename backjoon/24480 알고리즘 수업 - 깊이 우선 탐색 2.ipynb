```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)

N,M,R = map(int,input().split())

graph = [ [] for i in range(N+1)]

for i in range(M):
    u,v = map(int,input().split())
    graph[u].append(v)
    graph[v].append(u)
    
for i in range(1,N+1):
    graph[i].sort(reverse=True)
    
visited = [0]*(N+1)
cnt = [0]*(N+1)
count = 0

def dfs(v,e,r):
    global count
    visited[r] = 1
    count+=1
    cnt[r]=count
    for i in graph[r]:
        if visited[i] == 0:
            dfs(v,e,i)
dfs(N,M,R)

for i in range(1,N+1):
    print(cnt[i])
```
