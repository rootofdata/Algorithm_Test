```python
while True:
    n= int(input())
    if n == -1:
        break
    ans = []
    answer = ''
    for i in range(1,n):
        if n%i==0:
            ans.append(i)
    for i in ans:
        answer += str(i)
        answer += ' + '
        
    if n == sum(ans):
        print(n,'=',answer[:-3])
    else: print(n,'is NOT perfect.')
    
    
```
