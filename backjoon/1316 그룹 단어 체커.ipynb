```python
n = int(input())
count = 0
for i in range(n):
    a = input()
    if list(a) ==sorted(a,key=a.find):
        count+=1
print(count)
```

```python
N = int(input())
word_list = N

for i in range(N) :
    word = input()
    for j in range(len(word)-1) :
        if word[j] == word[j+1] :
            continue
        elif word[j] in word[j+1:] :
            word_list -= 1
            break
print(word_list)
```
