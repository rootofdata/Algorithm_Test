```python
N,M = map(int,input().split())
N_list =[]
M_list=[]
for i in range(N):
    N_list.append(list(map(int,input().split())))
    
M,K = map(int,input().split())
for j in range(M):
    M_list.append(list(map(int,input().split())))
    
ans = [[0]*K for _ in range(N)]

for i in range(N):
    for j in range(K):
        for l in range(M):
            ans[i][j] += (N_list[i][l]*M_list[l][j])
    
for i in range(N):
    print(*ans[i])
```
