```python
a,b= map(int,input().split())
c,d= map(int,input().split())
e,f = map(int,input().split())

x = sorted([a,c,e])
y = sorted([b,d,f])

if x[0]==x[1]:
    result_x=x[2]
else:
    result_x=x[0]
if y[0]==y[1]:
    result_y=y[2]
else:
    result_y=y[0]

print(result_x,result_y)
```
