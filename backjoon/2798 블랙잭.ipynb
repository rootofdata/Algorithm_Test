```python
# 3 
N,M = map(int,input().split())

a = list(map(int,input().split()))

# a  3   M    .
total = 0
for i in range(N-2):
    for j in range(i+1,N-1):
        for k in range(j+1,N):
            
            if a[i]+a[j]+a[k] <= M:
                total = max(total,a[i]+a[j]+a[k])
print(total)
```

```python
# combinations  
from itertools import combinations

N,M = map(int,input().split())
a = list(map(int,input().split()))
answer = 0

for comb in combinations(a,3):
    result = 0
    for c in comb:
        result +=c
    if result <=M:
        answer = max(result,answer)
print(answer)
```

```python
#  

N,M = map(int,input().split())
a = list(map(int,input().split()))
answer = 0
s=[]

def dfs(start):
    global answer
    if len(s)==3:
        result = sum(s)
        if result <=M:
            answer = max(answer,result)
        return answer
    for i in range(start,N):
        s.append(a[i])
        dfs(i+1)
        s.pop()
dfs(0)
print(answer)
```
