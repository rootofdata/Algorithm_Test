```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)

n,m,r=map(int,input().split())

graph = [ [] for i in range(n+1)]

for i in range(m):
    u,v = map(int,input().split())
    graph[u].append(v)
    graph[v].append(u)
    
for i in range(n):
    graph[i].sort()
    
visited = [0]*(n+1)
cnt = [0]*(n+1)

def dfs(v,e,r):
    global count
    visited[r]=1
    count+=1
    cnt[r]=count
    for i in graph[r]:
        if visited[i]==0:
            dfs(v,e,i)
count=0
dfs(n,m,r)

for i in range(1,n+1):
    print(cnt[i])
```
