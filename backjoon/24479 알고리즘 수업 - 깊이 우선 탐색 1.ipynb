```python
import sys
input = sys.stdin.readline
sys.setrecursionlimit(10**9)

#입력 
n,m,r=map(int,input().split())

#그래프 데이터. 
graph=[ [] for i in range(n+1)]

#입력, 그래프 데이터에 넣기
for i in range(m):
    u,v=map(int,input().split())
    graph[u].append(v)
    graph[v].append(u)

#오름차순으로 방문하기 정렬 (set 이용하면 안된다. set는 순서가 없다★)
for i in range(n):
    graph[i].sort() #★

#방문한 곳은 방문하지 않아야 한다
visited=[0]*(n+1)

#각각의 정점이 몇 번째로 방문되었는가. count값을 담아줌
cnt=[0]*(n+1)

#dfs
def dfs(v,e,r):
    global count #몇번째로 방문되었는지 세기 위함
    visited[r]=1
    count+=1
    cnt[r]=count
    #print("방문",r)
    for i in graph[r]:
        if visited[i]==0:    
            dfs(v,e,i)

#시작점으로 실행
count=0
dfs(n,m,r)

#각 몇번째로 방문되었는지 출력
for i in range(1,n+1):
    print(cnt[i])
    
# [출처] 백준 24479번 '알고리즘 수업 - 깊이 우선 탐색 1' 파이썬(Python) /DFS|작성자 inpink
```
