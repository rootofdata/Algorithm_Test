```python
# (1,1) 
## 5
## R R R U D D   --> 3 4
```

```python
n=int(input())
x,y = 1,1

# x :  ,  y : ,  
plans = input().split()
#L R U D
dx = [0,0,-1,1]
dy = [-1,1,0,0]
type = ['L','R','U','D']

for i in plans:
    for j in range(len(type)):
        if i == type[j]:
            nx = x+dx[j]
            ny = y+dy[j]
    if nx<1 or ny<1 or nx>n or ny>n:
        continue
    x,y = nx,ny
    
print(x,y)
```
